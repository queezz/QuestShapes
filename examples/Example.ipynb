```python
fig = gcf()
fig.set_size_inches([8,6])

coilfontsize = 10
ax = gca()

lcfs = np.loadtxt('inboardlimiter.txt')
#lcfs = np.loadtxt('inboardnull.txt')
#lcfs = np.loadtxt('divertor.txt')


quest = Quest(scale=-1)
cross = Cross(scale=1.5,origin=[-3,0])

plot(*quest.xy,'k')
plot(lcfs[:,0]*-1,lcfs[:,1])
plot(*cross.xy,'k')
plot([-4,0],[0,0],'C2--',zorder=0)

ax.set_aspect(1)
axis('off')


svpth = './images/QuestProbeSide.svg'
savefig(svpth,dpi=300,bbox_inches = 'tight')
```
